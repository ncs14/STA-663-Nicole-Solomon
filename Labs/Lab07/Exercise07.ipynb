### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probabbility of a rat dying. We are given $n = 5$, but we believve that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
rats = np.array([[-0.896,5,0],[-0.296,5,1],[-0.053,5,3],[0.727,5,5]])
rat_dat = pd.DataFrame(rats)
rat_dat.columns = ['dose','nrats','ndeaths']
rat_dat
```

```python
np.random.seed(1234)

rat_code = """
data {
    int<lower=0> N; // number of obs
    int<lower=0> n[N]; // number of rats
    int<lower=0> y[N]; // number of deaths
    real x[N]; // drug dose
}
transformed data {}
parameters {
    real a;
    real b;
}
transformed parameters {
    real<lower=0, upper=1> p[N];
    for (i in 1:N) {
        p[i] <- inv_logit(a + b*x[i]);
        }
}
model {
    a ~ normal(0, 5);
    b ~ normal(0, 10);
    y ~ binomial(n, p);
}
generated quantities {}
"""

# data must be fed as dictionary
rat_dat = {
             'N': 4,
             'n': [5,5,5,5],
             'y': [0,1,3,5],
             'x': [-0.896,-0.296,-0.053,0.727],
            }

# STAN can run multiple chains w/parallel processing
fit = pystan.stan(model_code=rat_code, data=rat_dat, iter=1000, chains=1)
```

### Using a hierarchcical model

This uses the Gelman radon data set and is based off this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the presence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
radon.head()
# radon = alpha_county + beta_county_(house w/wo basement)*floor
```

```python
np.random.seed(1234)

radon_code = """
data {
    int<lower=0> C; // num counties;
    int<lower=0> N; // num observations;
    int<lower=0> T[N]; // numeric factor version of county var;
    real x[N];
    real y[N]; 
}
transformed data {}
parameters {
    real alpha[C];
    real beta[C];
    
    real mu_a;
    real<lower=0> sigma_a;
    
    real mu_b;
    real<lower=0> sigma_b;
    
    real<lower=0> sigma_y;
}
transformed parameters {}
model {    
    mu_a ~ normal(0, 10000);
    sigma_a ~ uniform(0,100);
    
    mu_b ~ normal(0, 10000);
    sigma_b ~ uniform(0,100);
    
    alpha ~ normal(mu_a,sigma_a);
    beta ~ normal(mu_b,sigma_b);
    
    sigma_y ~ uniform(0,100);
    
    for (n in 1:N)
      y[n] ~ normal(alpha[T[n]] + beta[T[n]] * x[n], sigma_y);
}
generated quantities {}
"""

test = np.repeat(np.arange(len(radon.county.unique()))+1,np.array(radon.county.value_counts()))

radon_dat = {
             'N': radon.shape[0],
             'T': test,
             'C': len(radon.county.unique()),
             'x': radon.floor,
             'y': radon.log_radon
            }

fit = pystan.stan(model_code=radon_code, data=radon_dat, iter=1000, chains=1)
```

Implement the leap-frog method and simulate the spring dynamics following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the spring trajectory plots.

```python
# Note: mass = 1, spring constant = 1

def leap_frog(x1,p1):
    p2 = p1 - d*Up(x1)
    x2 = x1 + d*Kp(p2)
    return x2, p2

def U(x): return 0.5*x**2
def K(p): return 0.5*p**2
def Up(x): return x
def Kp(p): return p

### delta
d = 0.1


### Initial conditions
x0 = -4. # POSITION
p0 = 1.  # MOMENTUM


### Number of steps
L = 63

### Array to store x position and respective momentum
xpos = np.empty(L)
pmom = np.empty(L)
xpos[0] = x0
pmom[0] = p0


### Run leap-frog method for L steps in time

# perform the first half-step for updating momentum
p_new = p0 - d/2.*Up(x0)

# compute new position from first half-step
x_new = x0 + d*Kp(p_new)

for i in np.arange(L-1):
    if i in range(0,L,21):
        plt.show()
    
    # update position and momentum
    x_new, p_new = leap_frog(x_new,p_new)
    #xpos[i+1] = x_new
    #pmom[i+1] = p_new
    
    plt.subplots_adjust(hspace=0.5)
    
    # Oscillator plot
    ax1 = plt.subplot(211)
    plt.cla()
    xx = np.linspace(-6.,x_new,1000);
    plt.plot(xx,np.sin(6.*np.linspace(0,2*np.pi,1000)),'k-');
    plt.plot(x_new+.5,0.,'bo',markersize=20)
    plt.xlabel('x')
    ax1.set_xlim([-6,6]), ax1.set_ylim([-1,1])
    plt.title('Harmonic Oscillator')
    
    # Energy plot
    ax2 = plt.subplot(223)
    plt.cla()
    ax2.bar(0.25, U(x_new), 0.5, color='blue')
    ax2.bar(0.25, K(p_new), 0.5, color='yellow',bottom=U(x_new))
    ax2.bar(1, U(x_new)+K(p_new), 0.5, color='yellow')
    ax2.set_xticks((0.5,1.25))
    ax2.set_xticklabels(("U+K","H"))
    ax2.set_xlim([0.15,1.6]), ax2.set_ylim([0,10])
    plt.title("Energy")

#     bar_id = np.array([1,2])
#     bar1_vals = np.array([U(x_new),U(x_new)+K(p_new)])
#     bar2_vals = np.array([K(p_new),0.])
#     labels = ['U+K','H']
#     plt.bar(bar_id,bar1_vals,color='b',align='center')
#     plt.bar(bar_id,bar2_vals,color='y',bottom=np.array([U(x_new),0]),align='center')
#     plt.xticks(bar_id,labels)
#     ax2.set_ylim([0,10])
#     plt.title('Energy')
    
    # Phase plot
    ax3 = plt.subplot(224,aspect='equal')
    plt.cla()
    plt.plot(x_new,p_new,'ko',markersize=20)
    ax3.set_xlim([-6,6]), ax3.set_ylim([-6,6])
    plt.xlabel('x'), plt.ylabel('p')
    plt.title('Phase Space')

# complete the final momentum half-step
pmom[-1] = pmom[-1] - d/2.*xpos[-1]
```

**Exercise 4 (20 points)**: 

Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Basyesian Data Analysis" by Jouh Kruschke.) 

```python
np.random.seed(1234)

mu = np.array([0,0])
sigma = np.array([[1,0.8],[0.8,1]])

### Problem set-up:
# p(x)  = N_2(mu, sigma)
# U(x)  = -log(p(x)) == (x^T*sigma^(-1)*x)/2
# dU/dx = x
# K(p)  = p^Tp/2 ~ N(0,1)
# dK/dp = p


def U(x): return x.T.dot(np.linalg.inv(sigma)).dot(x)*0.5
def Up(x): return x.T.dot(np.linalg.inv(sigma))
def K(p): return p.T.dot(p)/2
def Kp(p): return p


### Steps and number of reps
d = 0.3
reps = 1000
L = 20


### Initial conditions
x = np.zeros(2*reps).reshape(2,reps);
x0 = np.array([0.,6])
x[:,0] = x0
#p = np.empty([2,reps])
 

### Begin HMC ###

#counter
t = 0

while t<(reps-1):
    
    t += 1
    
    # Initial momentum & position
    x0 = x[:,(t-1)]
    p0 = stats.norm.rvs(size=2)
    #if t==2: p[:,0] = p0

    ### Run leap-frog for L-steps w/start pt (x0,p0)

    # perform the first half-step for updating momentum
    p_star = p0 - d/2.*Up(x0)

    # compute new position from first half-step
    x_star = x0 + d*Kp(p_star)

    for i in np.arange(L-1):
        x_star, p_star = leap_frog(x_star,p_star)

    # complete the final momentum half-step
    p_star = p_star - d/2.*Up(x_star)
    # proposed state = x_star, p_star

    # compute Metropolis acceptance prob:
    acc = min(1, np.exp( -(U(x_star)+K(p_star)) - ( -(U(x0)+K(p0)) ) ) )

    # draw from uniform distn
    u = stats.uniform.rvs()

    # accept or reject
    if u<=acc: x[:,t] = x_star #, p[:,(t-1)] = x_star, p_star
    else: x[:,t] = x0 #, p[:,(t-1)] = x0, p0
        

```

```python
plt.title('Hamiltonian MC for Bivariate Normal')
plt.scatter(x[0,:],x[1,:])
plt.plot(x[0,:50],x[1,:50],'ro-')
plt.xlabel('x')
plt.ylabel('y')
```
