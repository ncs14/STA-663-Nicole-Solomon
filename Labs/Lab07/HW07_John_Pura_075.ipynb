### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probabbility of a rat dying. We are given $n = 5$, but we believve that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
log_reg_code = """
data {
    int<lower=0> N; 
    real dose[N];      //log dose
    int<lower=0> rats[N];      //num of rats
    int<lower=0> y[N];         //num of deaths
}
transformed data {}
parameters {
    real a;
    real b;
}
transformed parameters {}
model {
    a ~ normal(0, 5);
    b ~ normal(0, 10);
    for(i in 1:N) {
        y[i] ~ binomial(rats[i],inv_logit(a + b*dose[i]));
  }
}
generated quantities {}
"""

#Setup data
dose= np.array([-.896,-.296,-.053,.727])
rats = 5*np.ones(4,dtype=int)
deaths =np.array([0,1,3,5], dtype=int)

log_reg_dat = {
             'N': len(dose),
             'dose': dose,
             'rats': rats,
             'y': deaths
            }

fit = pystan.stan(model_code=log_reg_code, data=log_reg_dat, iter=2000, chains=1)
```

```python
res = fit.plot()
```

###From the above fit results, the mean value of b > 0, so the probability of a rat dying is positively associated with the drug dose. Furthermore, this relationship is strongly supported by the data as the 95% credible intervals (2.5% - 97.5%) do not include zero.

### Using a hierarchical model

This uses the Gelman radon data set and is based off this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the presence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
radon.head()
```

```python
from pandas import Categorical

#Convert categorical county names to integers
radon['county_id']= Categorical(radon.county).labels+1

#Convert float floor variable to integers
radon.floor = radon.floor.astype(int)

radon.head()
```

#### Model Setup

Let $a_c$ and $b_c$ be the intercept and slope for each county $c$, and assume they come from a independent group distributions

$$
a_c \sim \mathcal{N}(\mu_a, \sigma_a^2) \\
b_c \sim \mathcal{N}(\mu_b, \sigma_b^2)
$$

Also we assume that the hyperparameters come from the following distributions

$$
\mu_a \sim \mathcal{N}(0, 100^2) \\
\sigma_a \sim \mathcal{U}(0, 100) \\ 
\mu_b \sim \mathcal{N}(0, 100^2) \\
\sigma_b \sim \mathcal{U}(0, 100)
$$

Additionally, for between household errors $\epsilon$, we assume

$$
\epsilon \sim \mathcal{N}(0,\sigma^2) \\
\sigma^{2} \sim InvGamma(10,1)
$$

###We wish to test for the effect of basement floor on radon level through estimating $b_c$ and to see if there is substantial evidence of statistical interation between having a basement floor and county (i.e. if the effect of basement floor on radon level changes across counties).

```python
lin_reg_code = """
data {
    int<lower=0> N; 
    int<lower=1> L;            //L(evels) = 85
    int<lower=1,upper=L> c[N]; //county levels = 1-85
    int<lower=0,upper=1> f[N]; //floor variable
    real log_radon[N];
}
transformed data {}
parameters {
    real a[L]; //1D vector - 85 county levels
    real b[L];
    real mua;
    real mub;
    real<lower=0> sda;
    real<lower=0> sdb;
    real<lower=0> sigma;
}
transformed parameters {}
model {
    //hyperparameters
    mua ~ normal(0,100);
    sda ~ uniform(0,100);
    mub ~ normal(0,100);
    sdb ~ uniform(0,100);

    for (i in 1:L){    //L = county levels
        a[i] ~ normal(mua, sda);
        b[i] ~ normal(mub, sdb);
    }
    
    //Household errors
    sigma ~ gamma(10,1);
    
    for(j in 1:N) {
        log_radon[j] ~ normal(a[c[j]] + b[c[j]]*f[j],sigma);
    }
}
generated quantities {}
"""
 

lin_reg_dat = {
             'N': len(radon),
             'L': max(radon.county_id),
             'c': radon.county_id,
             'f': radon.floor,
             'log_radon': radon.log_radon
            }

fit = pystan.stan(model_code=lin_reg_code, data=lin_reg_dat, iter=1000, chains=1)
```

```python
#print fit
```

```python
b_res = pd.DataFrame(fit.extract(permuted=True)['b'][:,:]) 

b_mean = b_res.mean().values
b_lcl, b_ucl = b_res.quantile([0.025, 0.975]).values
```

```python
plt.plot(range(1,len(b_mean)+1),b_mean,'ko',
         range(1,len(b_mean)+1),b_lcl,'ro',
         range(1,len(b_mean)+1),b_ucl,'ro')
plt.axhline(0.0)

plt.ylabel('County-specific slope (arb. units)')
plt.xlabel('County')
plt.title('Posterior estimates of effect of basement floor on log-radon level')
```

###Since the data was coded as floor=0 as having a basement, the above plot indicates the average INCREASE in log-radon level (black dots) associated with HAVING a basement is about 0.5 units. The corresponding 95% credible intervals (red dots) indicate there is moderate to strong evidence (it includes zero, but just barely) that this increase is statistically important. Additionally, this graph suggests that the relationship between log-radon levels and having a basement is consistent across counties. That is, regardless of the county, having a basement is a risk hazard with regards to radon level accumulation.

**Exercise 4 (20 points)**: Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Basyesian Data Analysis" by Jouh Kruschke.) 

```python
#HYBRID MONTE CARLO SAMPLING -- BIVARIATE NORMAL
np.random.seed(12345)

# STEP SIZE
delta = 0.3
nSamples = 1000
L = 20

sigma = np.array([[1,0.8],[0.8,1]])
 
# DEFINE POTENTIAL ENERGY FUNCTION
def U(x,sigma):
    return np.dot(np.dot(x.T,np.linalg.inv(sigma)),x)

# DEFINE GRADIENT OF POTENTIAL ENERGY
def dU(x,sigma):
    return np.dot(x.T,np.linalg.inv(sigma))

# DEFINE KINETIC ENERGY FUNCTION
def K(p):
    return np.dot(p.T,p)/2.0
 
# INITIAL STATE
x = np.zeros((2,nSamples+1))
x0 = np.array([0.,6])
x[:,0] = x0.T
    
t = 0
while t < nSamples:
    t += 1
 
    # SAMPLE RANDOM MOMENTUM
    p0 = np.random.normal(size=(1,2))
 
    ## SIMULATE HAMILTONIAN DYNAMICS
    # FIRST 1/2 STEP OF MOMENTUM
    pStar = p0 - delta/2.*dU(x[:,t-1],sigma)
 
    # FIRST FULL STEP FOR POSITION/SAMPLE
    xStar = x[:,t-1] + delta*pStar
 
       
    # FULL STEPS
    for jL in range(L-1):
        # MOMENTUM
        pStar -= delta*dU(xStar.T,sigma)
        # POSITION/SAMPLE
        xStar += delta*pStar
    
    # LAST HALP STEP
    pStar -= delta/2.*dU(xStar.T,sigma)
 
    # EVALUATE ENERGIES AT
    # START AND END OF TRAJECTORY
    U0 = U(x[:,t-1],sigma)
    UStar = U(xStar.T,sigma)
    
    K0 = K(p0.T)
    KStar = K(pStar.T)

    
    pStar -= delta*0.5*dU(xStar.T,sigma)
    
    
    # ACCEPTANCE/REJECTION CRITERION
    alpha = min(1.0,np.exp((U0 + K0) - (UStar + KStar)))
 
    
    u = np.random.uniform(0,1)
    if (u < alpha):
        x[:,t] = xStar
    else:
        x[:,t] = x[:,t-1]

# DISPLAY
plt.figure(figsize=(5,5))
plt.scatter(x[0,:],x[1,:],c='k')
plt.plot(x[0,0:49],x[1,0:49],'ro-',linewidth=2,mfc='none',mec='r')
xlim = plt.xlim(-6, 6)
ylim = plt.ylim(-6, 7)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['1st 50 States','Samples'],loc=3)
plt.title('Hamiltonian Monte Carlo')
```
