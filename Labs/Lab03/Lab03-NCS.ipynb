**Exercise 1 (10 pts)**. Avoiding catastrophic cancellation.

The tail of the standard logistic distributon is given by $1 - F(t) = 1 - (1+e^{-t})^{-1}$.

- Define a function `f1` to calculate the tail probability of the logistic distribution using the formula given above
- Use `sympy` to find the exact value of the tail distribution (using the same symbolic formula) to 20 decimal digits
- Calculate the *relative error* of `f1` when $t = 25$ (The relative error is given by `abs(exact - approximate)/exact`)
- Rewrite the expression for the tail of the logistic distribution using simple algebra so that there is no risk of cancellation, and write a function `f2` using this formula. Calculate the *relative error* of `f2` when $t = 25$. 
- How much more accurate is `f2` compared with `f1` in terms of the relative error?

````python
##### 1. Function to calculate tail prob of logistic distn
""" Calculate upper tail Pr of logistic distn """
def f1(t):
    return 1. - pow(1. + np.exp(-t),-1)

x = 25.
y = f1(x)
print "Approximate upper tail prob at f1(t=25): ", y


##### 2. Function that eliminates risk of cancellation
def f2(t):
    return pow(1.+np.exp(t),-1)
y2 = f2(x)
print "Approximate upper tail at f2(t=25): ", y2


##### 3. Exact value of tail distn using sympy
y_exact = sympy.N(1. - pow(1. + sympy.exp(-25),-1),20)
print "\nExact value of upper tail prob: ", y_exact


##### 4. Relative error of f1
relerr = abs(y_exact - y)/y_exact
print "\nRelative error of f1: ", relerr


##### 5. Relative error of f2
relerr2 = abs(y_exact - y2)/y_exact
print "Relative error of f2: ", relerr2


##### 6. How much more accurate is f2?
print "\nf2 is more accurate by a factor of: ", round(relerr/relerr2)
````

**Exercise 3 (10 pts)**. Importance of using efficient algorihtms.

- Implement bubble sort
- Calculate its big $\mathcal{O}$ algorithmic complexity
- Time the performance of bubble sort on random uniform deviate vectors of sizes `range(100, 2000, 100)` using `time.time()` from the standard library
- Use `scipy.optimize.curve_fit` to fit an appropriate function to the collection of (size, execution time) data points. Extrapolate how long it would take to sort a random vector of size 1,000,000. Now time how long it takes for the system sort to sort a random vector of size 1,000,000.
- Plot the fits together with the data points using `matplotlib.pyplot` functions.

````python
def bubblesort(a):
    """ Function to perform bubble sort on list of numbers """
    # Run sort for the first (n-1), (n-2), ... sets of numbers
    for j in range(len(a)-1):
        # Run basic bubble sort for the given set of numbers
        for i in range(0,j):
            if a[i+1] < a[i]:
                a[i], a[i+1] = a[i+1], a[i]
    return a

# Test list of numbers
#test = [5,1,9,12,25,14,6,40,23,19]
# Test fcn
#print "Test bubblesort: \n", bubblesort(test)
#%timeit bubblesort(test)



#### Big O complexity: 
    # 2 for loops --> O(n^2)
 # n integers are checked ~ (n-1) times --> n*(n-1) == O(n^2)
print "The Big Oh complexity of bubblesort is O(n^2)"



#### Time performance of bubble sort on varying length vectors
# List of vector lengths
lengths = range(100,2000,100)
# list to hold times
times = list()
# Random seed to ensure replicability
np.random.seed(1110)
# Import time library
import time

for i in lengths:
    # initialize start time
    start = time.time()
    # sort random uniform vector of length lengths[i]
    bubblesort(np.random.uniform(size=i))
    # compute time to run sort = current time - start time
    timediff = (time.time() - start)
    # retain time
    times.append(timediff)

#### array to hold vector lengths and times
timearray = np.column_stack((lengths,times))

#print "\nLengths x Times: \n", timearray
````

````python
#### function to give to curve_fit to return predictive parameter estimates
def sort_O(x,a,b,c):
    """ Quadratic equation to obtain timing parameters """
    return a*x**2 + b*x + c

#### Run curve_fit to estimate predictive parameters
from scipy.optimize import curve_fit
estimates = curve_fit(sort_O,lengths,times)[0]
print "Exponential fit estimates: \n", estimates



#### Estimate the time to sort a vector of length 1M
oneM_sorttime = sort_O(1.0e+06,estimates[0],estimates[1],estimates[2])
print "\nTime to bubblesort vector of length 1M: ", round(oneM_sorttime,2), "secs or", round(oneM_sorttime/60/60/24,2), "days"
````

**Exercise 4 (20 pts)**. One of the goals of the course it that you will be able to implement novel algorithms from the literature. 

- Implement the mean-shift algorithm in 1D as described [here](http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/TUZEL1/MeanShift.pdf). 
    - Use the following function signature
    ```python
    def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    ```
    - xs is the data set, x is the starting location, and kernel is a kernel function
    - tol is the difference in $||x||$ across iterations
- Use the following kernels with bandwidth $h$ (a default value of 1.0 will work fine)
    - Flat - return 1 if $||x|| < h$ and 0 otherwise
    - Gaussian 
    $$\frac{1}{\sqrt{2 \pi h}}e^{\frac{-||x||^2}{h^2}}$$
    - Note that $||x||$ is the norm of the data point being evaluated minus the current value of $x$
- Use both kernels to find all 3 modes of the data set in `x1d.npy`
- Modify the algorithm and/or kernels so that it now works in an arbitrary number of dimensions.
- Use both kernels to find all 3 modes of the data set in `x2d.npy`
- Plot the path of successive intermediate solutions of the mean-shift algorithm starting from `x0 = (-4, 10)` until it converges onto a mode in the 2D data for each kernel. Superimpose the path on top of a contour plot of the data density.

````python
def normx(xs,x,h,num):
    """ Compute norm-1 or norm-2 of d-dim data """
    if np.ndim(xs)==1: # 1-dim case
        return (abs(np.subtract(x,xs))/h)**num # = |(x-x_i)/h|, ((x-x_i)/h)^2, ...
    else: # d-dim, d>2
        return np.linalg.norm(np.subtract(x,xs)/h,ord=num,axis=1)**num # same as above for >1-dim x
    
    
def flat_kern(xs,x,h):
    """ Flat kernel computation """
    return np.less(normx(xs,x,h,1),h).astype(int) # return 0/1 indicators for |(x-x_i)/h| < h


def gaus_kern(xs,x,h):
    """ Gaussian kernel computation """
    return 1/(2.*np.pi*h)**0.5 * np.exp(-normx(xs,x,h,2)/2.) # return Gaussian kernel of ((x-x_i)/h)^2


def mean_shift(xs, xi, kernel, h, toplot=False, max_iters=100, tol=1e-6):
    """ Mean shift computation and mode convergence function """
    # xs: data set, x: starting location, kernel: kernel function, max_iters: max iterations, tol: diff in ||x|| across iters
    
    # Initialize counter for iterations
    i = 1
    
    # Initialize array of converging points in the mean-shift solution
    plotpts = []
    
    # Continue looping until convergence on mode
    if np.ndim(xs)==1:
        x = xi
        # Initialize mean shift value
        mx = 1
        while abs(mx) > tol and i<=max_iters:
            # If toplot=True, retain search point
            if toplot:
                plotpts.extend(x)
            # Gaussian kernel
            if kernel in ["Gaussian","gaussian","gauss","gaus"]:
                g = gaus_kern(xs,x,h)
            # flat kernel
            else:
                g = flat_kern(xs,x,h)
            # compute mean shift
            mx = g.dot(xs)/sum(g) - x
            # if mean shift > tol then shift x
            if abs(mx) > tol: 
                x += mx
            i += 1
    else:
        x=xi.copy()
        # Initialize mean shift so as to start while loop
        mx = np.array([[1,1]])
        while np.linalg.norm(mx) > tol and i<=max_iters:
            # If toplot=True, retain search point
            if toplot:
                plotpts.extend(x)
            # Gaussian kernel
            if kernel in ["Gaussian","gaussian","gauss","gaus"]:
                g = gaus_kern(xs,x,h)
            # flat kernel
            else:
                g = flat_kern(xs,x,h)
            # compute mean shift
            mx = np.dot(xs.T,g)/g.sum() - x
            # if mean shift > tol then shift x
            if np.linalg.norm(mx) > tol:
                x += mx
            i += 1
    # return mode
    return x, np.array(plotpts).reshape(len(plotpts)/2,2)
````

````python
#### Load dataset
x1d = np.load('x1d.npy')

##################### Search entire dataset for modes #####################

#### X1d modes with flat kernel: 1,4,-5
x1modes_f = map(lambda y: round(mean_shift(xs=x1d,xi=y,kernel="flat",h=1.38)[0]), x1d) 
print "x1d modes with flat kernel: ", list(set(x1modes_f))

#### X1d modes with Gaussian kernel
x1modes_g = map(lambda y: round(mean_shift(xs=x1d,xi=y,kernel="gauss",h=0.75)[0]), x1d)
print "\nx1d modes with Gaussian kernel: ", list(set(x1modes_g))
````
