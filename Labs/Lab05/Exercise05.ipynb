Background for Exercises 1 and 2
----

The first 2 exercises are about using Newton's method to find the cube roots of unity - find $z$ such that $z^3 = 1$. From the fundamental theorem of algebra, we know there must be exactly 3 complex roots since this is a degree 3 polynomial.

We start with Euler's fabulous equation
$$
e^{ix} = \cos x + i \sin x
$$

Raising $e^{ix}$ to the $n$th power where $n$ is an integer, we get from Euler's formula with $nx$ substituting for $x$
$$
(e^{ix})^n = e^{i(nx)} = \cos nx + i \sin nx
$$

Whenever $nx$ is an integer multiple of $2\pi$, we have
$$
\cos nx + i \sin nx = 1
$$

So
$$
e^{2\pi i \frac{k}{n}}
$$
is a root of 1 whenever $k/n = 0, 1, 2, \ldots$.

So the cube roots of unity are $1, e^{2\pi i/3}, e^{4\pi i/3}$. 

While we can do this analytically, the idea is to use Newton's method to find these roots, and in the process, disover some rather perplexing behavior of Newton's method.

**Exercise 1 (10 points)**. Netwon's method for functions of complex variables - stability and basins of attraction.

1. Write a function with the following function signature `newton(z, f, fprime, max_iter=100, tol=1e-6)` where
    - `z` is a starting value (a complex number e.g.  ` 3 + 4j`)
    - `f` is a function of `z`
    - `fprime` is the derivative of  `f`
The function will run until either max_iter is reached or the absolute value of the Newton step is less than tol. In either case, the function should return the number of iterations taken and the final value of `z` as a tuple (`i`, `z`). 

2. Define the function `f` and `fprime` that will result in Newton's method finding the cube roots of 1. Find 3 starting points that will give different roots, and print both the start and end points. 

```python
def newton(z, f, fprime, max_iter=100, tol=1e-6):
    """ Use Newton's method to compute cube roots """
    # initialize number of iterations
    num_iter = 0
    
    # initialize step size
    step = 1
    
    while num_iter <= max_iter and abs(step) > tol:
        # step on path to discover root: f(z_(n-1)) / f'(z_(n-1))
        step = f(z)/fprime(z)
        # new path value = z_(n-1) - step
        z -= step
        num_iter += 1
    
    return num_iter, z

#### Test function
#def y(x):
#    return x**3 - 2*x**2 - 11*x +12
#def y1(x):
#    return 3*x**2 - 4*x - 11

#print newton(2.349,y,y1,100,1e-6)
```

**Exercise 2 (10 points)**. Write the following two plotting functions to see some (pretty) aspects of Newton's algorithm in the complex plane.

1. The first function `plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv')` calculates and stores the number of iterations taken for convergence (or max_iter) for each point in a 2D array. The 2D array limits are given by `extent` - for example, when `extent = [-1,1,-1,1]` the corners of the plot are `(-1, -i), (1, -i), (1, i), (-1, i)`. There are `n` grid points in both the real and imaginary axes. The argument `cmap` specifies the color map to use - the suggested defaults are fine. Finally plot the image using `plt.imshow` - make sure the axis ticks are correctly scaled. Make a plot for the cube roots of 1.

2. The second function `plot_newton_basins(f, fprime, n=200, extent=[-1,1,-1,1], cmap='jet')` has the same arguments, but this time the grid stores the identity of the root that the starting point converged to. Make a plot for the cube roots of 1 - since there are 3 roots, there should be only 3 colors in the plot.

```python
def plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv'):
    """ Plot the no. iterations needed to locate each cubed root of 1 in (-1,-i)x(1,i) grid """
    # vectorize the Newton fcn so as to handle array inputs
    newt_vec = np.vectorize(newton)
    
    # Define the arrays based on the grid margins == 'extent'
    X1, X2 = np.linspace(extent[0],extent[1],n), np.linspace(extent[2],extent[3],n)
    
    # Compute the iterations
    iters = newt_vec(X1[:,None]+1j*X2[None,:],f,fprime)[0].T
    
    # plot the number of iterations for each coordinate
    fig, ax = plt.subplots()
    im = ax.imshow(iters, cmap=plt.get_cmap(cmap), vmin=abs(iters).min(), vmax=abs(iters).max(), extent=extent, origin='right')
    cbar = plt.colorbar(im)
    cbar.set_label('# of iterations', rotation=270, labelpad=20)
    plt.title('Iterations for cubed roots of 1',y=1.05)
    ax.set_xlabel('Real')
    ax.set_ylabel('Imaginary')

plot_newton_iters(f1,f1p)
```

**Exercise 3 (10 points)**. Consider the following function on $\mathbb{R}^2$:

$$f(x_1,x_2) = -x_1x_2e^{-\frac{(x_1^2+x_2^2)}{2}}$$

1. Use `sympy` to compute its gradient.
2. Compute the Hessian matrix. 
3. Find the critical points of $f$.
4. Characterize the critical points as max/min or neither. Find the minimum under the constraint 
$$g(x) = x_1^2+x_2^2 \leq 10$$
and 
$$h(x) = 2x_1 + 3x_2 = 5$$ using `scipy.optimize.minimize`.
5. Plot the function using `matplotlib`.

```python
import sympy as sp
x1, x2 = sp.symbols('x1 x2')

#### Define the symbolic bivariate fcn
f2 = -x1*x2*exp(-(x1**2 + x2**2)/2)

#### Use sympy to compute gradient of f
print "Gradient of f:"
sp.Matrix([[sp.diff(f2,x1)],[sp.diff(f2,x2)]])
```

```python
#### Use sympy to compute Hessian matrix of f
print "Hessian matrix:"
sp.hessian(f2,(x1,x2))
```

```python
#### Obtain derivatives of f wrt x1, x2
dx1 = sp.diff(f2,x1)
dx2 = sp.diff(f2,x2)

#### Solve for the critical points of f
print "Critical points of f:"
cps = sp.solve([dx1,dx2],[x1,x2])
cps
```

```python
#### Characterize the cp's as min/min/neither
types = []
for i in range(len(cps)):
    type = sp.diff(sp.diff(f2,x1), x1).subs( {x1:cps[i][0], x2:cps[i][1]} )
    if type>0:
        types.append('min')
    elif type<0:
        types.append('max')
    else:
        types.append('neither')
print "Types of critical points:"
types
```

```python
from scipy import optimize as opt

# Make f2 a real (non-symbolic) function
def f2(x): return -x[0]*x[1]*np.exp(-(x[0]**2 + x[1]**2)/2)

# Set the constraints
cons = ({'type': 'eq',
         'fun' : lambda x: np.array([2*x[0] + 3*x[1] - 5]),
         'jac' : lambda x: np.array([2,3])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([x[0]**2 + x[1]**2 - 10]),
         'jac' : lambda x: np.array([2*x[0], 2*x[1]])})

f2_opt = opt.minimize(f2,[2,2],constraints=cons,method='L-BFGS-B')

# x_min at which fcn takes local min
minf2_x = f2_opt['x']
# fcn value at x_min
minf2 = round(f2(minf2_x),3)

print "Under constraints, local min(f2) =", minf2, "at (x1,x2) =", minf2_x
```

**Exercise 4 (20 points)**. Find the maximum likelihood function for a logistic regression. Load the file "train.csv" (located in this directory) into python.  This file is data from the survival data from the Titanic (from https://www.kaggle.com/).  You are to use a logistic regression to model survival as a function of gender, age and class (of travel). Find the maximum likelihood estimator of $\beta$ by numerical optimization using stochastic gradient descent as follows:

1. Stochastic gradient descent.  In this method, gradient descent is used essentially by fitting *one data point at a time*.  Recall the usual gradient descent step:

$$\beta_{i+1} = \beta_i - \nabla \ell(\beta_i)$$

where

$$\nabla \ell(\beta_i) = \sum\limits_{j=1}^n \nabla\ell(\beta_i,x_j,y_j)$$

and $\ell$ is the log-likelihood function.  All of the data is used to make the next step toward the optimal $\beta$.  In stochastic gradient descent, only one point at a time is used to determine the next $\beta$:

$$\beta_{i+1} = \beta_i - \alpha\nabla \ell(\beta_i,x_j,y_j)$$

where $\alpha$ is the step size.  For simplicity, we'll take a constant $\alpha=1$.

Implement the following stochastic gradient algorithm:
 
   a. Shuffle data points (i.e. randomly permute the order of the $(x_j,y_j)$
   
   b. Refine beta using the iterative formula above over each data point. 
   
   c. Repeat a and b until convergence is reached.
 
Apply the algorithm to the given data set to find the best-fit logistic regression coefficients.  Do not forget that your optimization should include a tolerance and a limit on the number of iterations.

```python
#### Read in the dataset train
train = pd.read_csv('train.csv')
train.shape
    # 891 observations total


#### Check var types & drop missing values
#train.info()
train2 = train.loc[:,['Survived','Pclass','Sex','Age']].dropna()
train2.shape
    # 714 observations remaining


#### Convert gender to numerical indicator
train2['Sex'] = (train2['Sex']=='male')*1


#### Convert Pclass to dummy vars
dummies = pd.get_dummies(train2['Pclass'])
# only need two of these
dummies.drop(1, axis=1, inplace=True)


#### Recompile into dataset with vars of interest
tr = pd.concat([train2, dummies], axis=1)      
tr.drop(['Pclass'], inplace=True, axis=1)
tr.columns = ['surv','sex','age','c2','c3']
tr.insert(1, 'int', 1)
#trdat = np.insert(tr.values,1,1,axis=1) # converts df to array

#### Convert to numpy array for use in function
trmat = tr.values
```

```python
def log_reg(dat,max_iters=10000,tol=1e-6):
    """ Compute MLE(Beta) using Stochastic Gradient Descent """
    # duplicate the data
    dat2 = dat.copy()
    
    # initialize betas, iter count, diff btwn Beta_i+1, Beta_i == gradient
    beta = np.zeros(dat.shape[1]-1)
    count = 0
    
    # set alpha
    a = 0.0001
    
    # set seed
    np.random.seed(1234)
    
    #Compute initial log likelihood = sum_i y_i^T * x_i * Beta^T - sum_i log(1+exp(x_i * Beta^T))
    ll0 = np.dot(dat2[:,0].T,np.dot(dat2[:,1:],beta.T)) - np.sum(np.log(1+np.exp(np.dot(dat2[:,1:],beta.T))))
    ll1 = float('inf')
    
    while abs(ll1 - ll0) >= tol or count <= max_iters:
        # shuffle data
        np.random.shuffle(dat2)
        
        # step through each observation
        for i in range(dat2.shape[0]):
            # select ith row to compute the gradient with
            x, y = dat2[i,1:], dat2[i,0]
            # compute the success prob
            p = np.exp(np.dot(x,beta.T))/(1+np.exp(np.dot(x,beta.T)))
            # compute the gradient
            grad = x*(y-p)
            # compute the new beta
            beta += a*grad
        
        # compute new log likelihood
        ll1 = np.dot(dat2[:,0].T,np.dot(dat2[:,1:],beta.T)) - np.sum(np.log(1+np.exp(np.dot(dat2[:,1:],beta.T))))
        ll0 = ll1
        
        count += 1
    
    return beta
```

```python
beta_est = log_reg(trmat)
```

```python
betaest = pd.DataFrame(beta_est).T
betaest.columns = ['Intercept','Sex','Age','Class2','Class3']
betaest.index = ["Beta_hat"]
print "Parameter estimates: \n", betaest
```
